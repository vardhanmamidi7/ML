```python
import numpy as np
import matplotlib.pyplot as plt
#derivative of loss function
def f(x):
    return x**2
def df(x):
    return 2*x

x_start=9
learning_rate=0.15
n_iterations=10


#for storing the iteration steps for visualization
x_steps=[x_start]
y_steps=[f(x_start)]

#values for plotting the parabola
x=np.linspace(-10,10,400)
y=f(x)

for i in range(n_iterations):
    plt.figure(figsize=(8,5))
    # plot the parabola
    plt.plot(x,y,label='f(x)=x^2')

    # current position on the parabola
    current_x=x_steps[-1]
    current_y=y_steps[-1]
    
    gradient=df(current_x)


    #Update the position (gradient descent step)

    next_x=current_x-learning_rate*gradient
    next_y=f(next_x)

    x_steps.append(next_x)
    y_steps.append(next_y)



    #plot the current steps path
    plt.scatter(x_steps,y_steps,color="green")
    plt.plot(x_steps,y_steps,linestyle="-",marker='o',color="blue",label="gradient descent path")


    #plot the tangent line (derivative at current point)
    tangent_x=np.linspace(current_x-1,current_x+1,10)
    tangent_y=gradient*(tangent_x-current_x)+current_y
    plt.plot(tangent_x,tangent_y,"--",color="red",linewidth=0.9,label="tangent at current step")
    plt.title(f"Gradient descent on f(x)=x^2 iteration{i+1}")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function f(x,y) = x^2 + y^2
def f(x, y):
    return x**2 + y**2

# Partial derivatives
def df_dx(x, y):
    return 2*x

def df_dy(x, y):
    return 2*y

# Gradient descent parameters
learning_rate = 0.1
n_iterations = 20
start_x, start_y = 8, 8

# Store history of points
history_x = [start_x]
history_y = [start_y]
history_z = [f(start_x, start_y)]

# Create meshgrid for plotting the surface 
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Perform gradient descent
current_x, current_y = start_x, start_y
for i in range(n_iterations):
    # Calculate gradients
    grad_x = df_dx(current_x, current_y)
    grad_y = df_dy(current_x, current_y)
    
    # Update position
    current_x = current_x - learning_rate * grad_x
    current_y = current_y - learning_rate * grad_y
    
    # Store points for visualization
    history_x.append(current_x)
    history_y.append(current_y)
    history_z.append(f(current_x, current_y))

# Create 3D visualization
fig = plt.figure(figsize=(12, 5))

# Surface plot
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax1.plot(history_x, history_y, history_z, 'r.-', linewidth=2, markersize=10, label='Gradient descent path')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('f(X,Y)')
ax1.set_title('Gradient Descent on f(x,y) = x + y')
fig.colorbar(surf)

# Contour plot
ax2 = fig.add_subplot(122)
contours = ax2.contour(X, Y, Z, levels=20)
ax2.clabel(contours, inline=True, fontsize=8)
ax2.plot(history_x, history_y, 'r.-', linewidth=2, markersize=10, label='Gradient descent path')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Contour Plot with Gradient Descent Path')
ax2.legend()

plt.tight_layout()
plt.show()

# Print final result
print(f"Starting point: ({start_x}, {start_y})")
print(f"Final point: ({history_x[-1]:.4f}, {history_y[-1]:.4f})")
print(f"Final value: {history_z[-1]:.4f}")
```
