```python
positive_points=np.array([(-2,3),(0,1),(2,-1)])  #truevalue=+1
negative_points=np.array([(-2,1),(0,-1),(2,-3)])# true value=-1
```

```python
x=np.vstack((positive_points,negative_points))
y=np.array([1]*len(positive_points)+[-1]*len(negative_points))
```

```python
def plot_decision_boundary(theta,theta_not,iteration):
    plt.figure(figsize=(10,8))
    plt.scatter(positive_points[:,0],positive_points[:,1],color="blue" ,label="positive_class")
    plt.scatter(negative_points[:,0],negative_points[:,1],color="orange" ,label="negative_class")

    x_values=np.linspace(-3,3,2)
    if theta[1] != 0:
        y_values=-(theta[0] *x_values+theta_not)/theta[1]
        plt.plot(x_values,y_values,label=f"iteration{iteration}")
    else:
        x_val=-theta_not/theta[0]
        plt.axvline(x=x_val,label=f"iteration{iteration}")

    plt.xlim(-3,3)
    plt.ylim(-4,4)
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title(f"decision_Boundary{iteration} ")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
theta_final,theta_not_final= perceptron(x,y,10)
plot_decision_boundary(theta_final, theta_not_final,1) 
```
